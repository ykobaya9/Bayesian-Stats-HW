```python
# Code for Bayesian Statistics HW1 - Yuta Kobayashi

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import beta
```

```python
# Problem 3 part 2

theta = np.arange(0, 1.01, 0.1)
y = 57

p_y_theta = (theta ** y) * ((1 - theta) ** (100 - y))

plt.figure()
plt.scatter(theta, p_y_theta)
plt.xlabel("theta")
plt.ylabel("p(y|theta)")
```

```python
# Problem 3 part 3

theta = np.arange(0, 1.01, 0.1)
y = 57

p_y_theta = (theta ** y) * ((1 - theta) ** (100 - y))
p_theta_y = (1/11) * p_y_theta

# compute normalizing constant z
sum = np.sum(p_theta_y)
z = 1 / sum

plt.figure()
plt.scatter(theta, p_theta_y * z)
plt.xlabel("theta")
plt.ylabel("p(theta|y)")
```

```python
# Problem 3 part 4

theta = np.linspace(0, 1, 100)
y = 57

p_y_theta = (theta ** y) * ((1 - theta) ** (100 - y))
p_theta_y = 1 * p_y_theta

plt.figure()
plt.plot(theta, p_theta_y)
plt.xlabel("theta")
plt.ylabel("p(theta|y)")
```

```python
# Problem 3 part 5

theta = np.linspace(0, 1, 100)
p_theta_y = beta(58, 44)

plt.figure()
plt.plot(theta, p_theta_y.pdf(theta))
plt.xlabel("theta")
plt.ylabel("p(theta|y)")
```

```python
# Problem 4

y = 57

x = np.linspace(0, 1, 101)

theta_0 = np.arange(0.1, 1.0, 0.1)
n_0 = np.array([1, 2, 8, 16, 32])

contour_z = []

for theta in theta_0:
  for n in n_0:
    a = theta * n
    b = (1 - theta) * n

    posterior = beta(a + y, b + 100 - y)
    cdf = posterior.cdf(x)

    p = 1 - cdf[50]

    contour_z.append(p)

contour_z = np.array(contour_z).reshape(len(theta_0), len(n_0))

contour_x, contour_y = np.meshgrid(n_0, theta_0)

plt.figure(figsize=(10, 10))
plt.contourf(contour_x, contour_y, contour_z)
plt.xlabel("n_0")
plt.ylabel("theta_0")
plt.colorbar();
```
